```
import requests
url = 'https://www.teleantioquia.co/noticias/'
response = requests.get(url)
print(response)
```

```
print (response.headers)
print(response.headers['Date'])
print(response.text)
```

```
import bs4
completo = bs4.BeautifulSoup(response.text,'html.parser')
noticias = completo.select('.tab-pane a')
print (noticias)
```

```
titulos =[]
links = []
for noticia in noticias:
  titulos.append(noticia.text)
  links.append( noticia['href'])
print (titulos)
print (links)
dictNoticias = {}
dictNoticias['Noticias']= titulos
dictNoticias['links']= links
```

```
import pandas as pd 
noticiasElTiempo = pd.DataFrame(dictNoticias)
noticiasElTiempo

noticiasElTiempo.to_csv('noticias.csv', index=False)
```
