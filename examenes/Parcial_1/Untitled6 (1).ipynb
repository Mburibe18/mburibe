```
import requests
url = 'https://listado.mercadolibre.com.co/computadoras-portatiles#D[A:computadoras%20portatiles]'
response = requests.get (url)
print (response)
```

```
import bs4
parseo = bs4.BeautifulSoup (response.text, 'html.parser')
getDataSelect = lambda clase : parseo.select(clase)
dataTitle = parseo.select ('.ui-search-item__title')
dataPrice = getDataSelect('.price-tag-fraction')
print (dataTitle)
print (dataPrice)
nombreComputadores2 = []
nombreComputadores = []
funcionExtraerTexto = lambda elemento : elemento.text
funcionListaDatos = lambda funcion, data :  list (map (funcion,data))
nombreComputadores = funcionListaDatos(funcionExtraerTexto, dataTitle)
for elemento in dataTitle :
  nombreComputadores2.append (elemento.text)

print (nombreComputadores)
print (nombreComputadores)
listaPrecios = funcionListaDatos(funcionExtraerTexto, dataPrice)
print (listaPrecios)
```

```
print(len(nombreComputadores))
print(len(listaPrecios))
listaPrecios= listaPrecios[:50]
print(len(listaPrecios))
```

```
import pandas as pd
dictComputadores = {}
dictComputadores ['Titulos'] = nombreComputadores
print (len (nombreComputadores), len (listaPrecios))
dictComputadores ['Precios'] = listaPrecios
computadoresDataFrame = pd.DataFrame(dictComputadores)
computadoresDataFrame.to_csv('Computadores.csv', index=False)
```

```
funcionAgregarComputador = lambda elemento : 'computador : ' + elemento
titulosComputadores = list (map(funcionAgregarComputador,nombreComputadores))
print (titulosComputadores)
titulosComputadores2 = funcionListaDatos (funcionAgregarComputador, nombreComputadores)
print (titulosComputadores2)
dictComputadores = {}
dictComputadores['Titulos'] = titulosComputadores2
dictComputadores ['Precios'] = listaPrecios
computadoresDF = pd.DataFrame(dictComputadores)
computadoresDF.to_csv('computadoresMod.csv', index= False) 
```

```
lenovo = lambda elemento : 'Lenovo' in elemento 
dell = lambda elemento : 'Dell' in elemento

listaLenovos = list (filter(lenovo, titulosComputadores))
listaDell = list (filter(dell, titulosComputadores))
print (listaLenovos)
print (listaDell)

nombresCortos = lambda elemento : len (elemento) < 80
listaLenovosFiltrados = list (filter (nombresCortos, listaLenovos))
print (listaLenovosFiltrados)

lenovoSerie = pd.Series(listaLenovos)
lenovoSerie.to_csv('Computadores lenovo.csv',  index_label='Lenovo')
dellSerie = pd.Series(listaDell)
dellSerie.to_csv('Computadores dell.csv',  index_label='Dell')
```
